##Parte I.- Operaciones aritméticas

####Ejercicio 1: Suma de imágenes.


```python
# Conexión con Google Drive para cargar imágenes
from google.colab import drive
drive.mount('/content/drive')

# Importación de librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Rutas de las imágenes
ruta1 = '/content/drive/MyDrive/imagenesLabPDI/brazo.jpg'
ruta2 = '/content/drive/MyDrive/imagenesLabPDI/torax.jpg'

# Lectura de imágenes en escala de grises
img1 = cv2.imread(ruta1, 0)
img2 = cv2.imread(ruta2, 0)

# Verificación de que sí cargaron
if img1 is None or img2 is None:
    print("Error al cargar imágenes. Revisa las rutas.")
else:
    print("Imágenes cargadas correctamente")

# Verificación de tamaños
print("Imagen 1:", img1.shape)
print("Imagen 2:", img2.shape)

# Ajuste de tamaño si es necesario
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Suma ponderada
suma_ponderada = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

# Suma promedio
suma_promedio = (img1 + img2) / 2
suma_promedio = suma_promedio.astype(np.uint8)

# Muestra de los resultados
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title('Imagen 1')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title('Imagen 2')
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title('Suma Ponderada')
plt.imshow(suma_ponderada, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title('Suma Promedio')
plt.imshow(suma_promedio, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

####Ejercicio 2: Resta de imágenes.


```python
# Resta normal
resta = cv2.subtract(img1, img2)

# Resta absoluta (evita negativos)
resta_abs = cv2.absdiff(img1, img2)

# Mostrar resultados
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.title('Imagen 1')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Resta normal')
plt.imshow(resta, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Resta Absoluta')
plt.imshow(resta_abs, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

####Ejercicio 3: Multiplicación por un factor.

```python
# Factores de brillo
factores = [0.5, 1.5, 2.0]

plt.figure(figsize=(12,8))

# Imagen 1
for i, factor in enumerate(factores):
    img1_mod = cv2.convertScaleAbs(img1, alpha=factor, beta=0)

    plt.subplot(2,4,i+1)
    plt.title(f'Img1 Factor {factor}')
    plt.imshow(img1_mod, cmap='gray')
    plt.axis('off')

# Original imagen 1
plt.subplot(2,4,4)
plt.title('Img1 Original')
plt.imshow(img1, cmap='gray')
plt.axis('off')

# Imagen 2
for i, factor in enumerate(factores):
    img2_mod = cv2.convertScaleAbs(img2, alpha=factor, beta=0)

    plt.subplot(2,4,i+5)
    plt.title(f'Img2 Factor {factor}')
    plt.imshow(img2_mod, cmap='gray')
    plt.axis('off')

# Original imagen 2
plt.subplot(2,4,8)
plt.title('Img2 Original')
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

####Ejercicio 4: Multiplicación elemento por elemento.

```python
# Convertir a float para evitar saturación
img1_float = img1.astype(np.float32)
img2_float = img2.astype(np.float32)

multiplicacion = img1_float * img2_float

# Normalizar para visualizar
multiplicacion = cv2.normalize(multiplicacion, None, 0, 255, cv2.NORM_MINMAX)
multiplicacion = multiplicacion.astype(np.uint8)

# Mostrar resultados
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.title('Imagen 1')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Imagen 2')
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Multiplicación')
plt.imshow(multiplicacion, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
