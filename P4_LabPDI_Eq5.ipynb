```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('mano.jpg', 0)

plt.imshow(img, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')
```

```python
#Parte 1. Procesamiento geométrico
#Ejercicio 1. Traslación
rows, cols = img.shape

M1 = np.float32([[1, 0, 50],
                 [0, 1, 30]])

traslacion1 = cv2.warpAffine(img, M1, (cols, rows))

M2 = np.float32([[1, 0, 20.5],
                 [0, 1, 15.5]])

traslacion2 = cv2.warpAffine(img, M2, (cols, rows))

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(traslacion1, cmap='gray')
plt.title('Traslación (50,30)')
plt.subplot(1,3,3)
plt.imshow(traslacion2, cmap='gray')
plt.title('Traslación (20.5,15.5)')
plt.show()
```

```python
#Ejercicio 2. Rotación
rows, cols = img.shape

centro = (cols/2, rows/2)

M = cv2.getRotationMatrix2D(centro, 45, 1)

rotada = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(rotada, cmap='gray')
plt.title('Rotación a 45°')
plt.show()
```

```python
#Ejercicio 3. Escala
# Escala 150%
escala_grande = cv2.resize(img, None, fx=1.5, fy=1.5)

# Escala 50%
escala_peque = cv2.resize(img, None, fx=0.5, fy=0.5)
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(escala_grande, cmap='gray')
plt.title('150%')
plt.subplot(1,3,3)
plt.imshow(escala_peque, cmap='gray')
plt.title('50%')
plt.show()
```

```python
#Parte 2. Ecualización de histograma
img2 = cv2.imread('RML.jpg', 0)

# Histograma original
plt.hist(img2.ravel(), bins=256)
plt.title('Histograma Original')
plt.show()
```

```python
ecualizada = cv2.equalizeHist(img2)
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img2, cmap='gray')
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(ecualizada, cmap='gray')
plt.title('Ecualizada')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(img2.ravel(), bins=256)
plt.title('Histograma Original')

plt.subplot(1,2,2)
plt.hist(ecualizada.ravel(), bins=256)
plt.title('Histograma Ecualizado')

plt.show()
```

##**CONCLUSIÓN**

A partir de esta práctica se comprobó que las técnicas de procesamiento digital de imágenes permiten modificar y mejorar significativamente la información visual. Las transformaciones geométricas, como la traslación, rotación y escalamiento, resultan útiles para cambiar la posición, orientación y tamaño de la imagen, manteniendo su estructura general. Sin embargo, también se observó que estas operaciones pueden generar zonas sin información y ligeras alteraciones en la calidad debido a la interpolación.

Por otro lado, la ecualización de histograma demostró ser una herramienta eficaz para mejorar el contraste, permitiendo una mejor visualización de los detalles al redistribuir los niveles de intensidad. Aun así, este proceso puede incrementar el ruido en ciertas regiones de la imagen.

En conjunto, los resultados obtenidos permiten entender cómo distintas técnicas afectan tanto la apariencia como la información contenida en una imagen, lo cual es fundamental para su análisis en aplicaciones reales.
